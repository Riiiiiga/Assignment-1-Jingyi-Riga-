In this step, we're dealing with the English text.First, we define the file path 'ENG.txt' and we're open the file specified by file_path in read mode ('r') with UTF-8 encoding. Next, we tokenize the data, we first used the regular expressions “re.findall” to extract words that consist of lowercase letters only. We used the regular expression r'\b[a-z]+\b' to match all the words with lowercase alphabets, and we obtained a list of words converted to lowercase. Next, we removed the stopwords and punctuations and the meaningless single letters except “I” and “a”; Afterwards, we loaded the language models for both languages, and we processed the text by joining them into a space-separated string, so that we can proceed to calculate the word frequency in the next steps.

```python
file_path = '/workspaces/computational-syntax/ESP.txt'
with open(file_path, 'r', encoding='utf-8') as my_file: 
    text = re.findall(r'\b[a-z]+\b',my_file.read().lower())  
    
nlp = spacy.load("en_core_web_sm")  
doc = nlp(" ".join(text))  

words = [token.text for token in doc if not token.is_stop and not token.is_punct and len(token.text) > 1 or token.text in ['a','i']]     
```

Now we are going to extract the target words and calculate the word frequency and their length.First,we create a 'counter' object from the list of words,and we retrieve the 5 most common words along with their frequencies from 'word_freq'. We print the title 'Top 5 common words', and then we ask to iterates over each word-frequency pair in 'common_words'.Inside this loop, we print each word with its corresponding frequency and length.
Similarly, we retrieves the 5 least common words from word_freq,which is done by slicing the list returned by most_common() in reverse order. We also print the title and with another loop that iterates over each word_frequency pair in least common words, we extract each word with its correspoding frequency and its length.

```python
word_freq = Counter(words)    
common_words = word_freq.most_common(5) 
print("Top 5 common words:")   
for word, frequency in common_words:      
    print(f"{word}: {frequency} times, Length: {len(word)}")   
least_common_words = word_freq.most_common()[:-6:-1]    
print("\nTop 5 uncommon words:")   
for word, frequency in least_common_words:  
    print(f"{word}: {frequency} times, Length: {len(word)}")
```

For the Spanish text, we also performed the same commands and we got the list of 5 most common words and 5 uncommon words.

```python
file_path='/workspaces/computational-syntax/ESP.txt'    
with open(file_path, 'r', encoding='utf-8') as my_file:   

    text = re.findall(r'\b[a-z]+\b',my_file.read().lower())  

    nlp = spacy.load("en_core_web_sm")
    doc = nlp(" ".join(text))      

    words = [token.text for token in doc if not token.is_stop and not token.is_punct] 
    word_freq = Counter(words)    
    common_words = word_freq.most_common(5)   
    print("Top 5 common words:")   
    for word, frequency in common_words:       
        print(f"{word}: {frequency} times, Length: {len(word)}")   
    least_common_words = word_freq.most_common()[:-6:-1]   
    print("\nTop 5 uncommon words:")  
    for word, frequency in least_common_words: 
        print(f"{word}: {frequency} times, Length: {len(word)}")  
```
